```python
#Libraries
import gradio as gr
from transformers import ViltProcessor, ViltForQuestionAnswering
from PIL import Image
```

```python
processor= ViltProcessor.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
model= ViltForQuestionAnswering.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
```

```python
def answer_question(image, text):
  try:
    encoding= processor(image, text, return_tensors="pt")
    outputs= model(**encoding)
    logits= outputs.logits

    idx= logits.argmax(-1).item()
    answer= model.config.id2label[idx]

    return answer

  except Execption as e:
    print(f"error: {e}")
    return "Sorry, I had trouble answering your question. Try a different image or ask another question."
```

```python
#Creating gradio web interface.
iface= gr.Interface(
    fn= answer_question,
    inputs=[
        gr.Image(type="pil"),
        gr.Textbox(label="Ask a question about the image...")],

    outputs= gr.Textbox(label="Answer"),
    title= "Muiltimodel AI: Visual Quesion Answering",
    description= "Upload an image and ask any question about it."
)

iface.launch()
```
